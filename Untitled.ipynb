# BANK MARKETING DECISION TREE CLASSIFIER

## INTRODUCTION

In this project, we aim to build a Decision Tree classifier to predict whether a customer will purchase a product or service based on their demographic and behavioral data. To achieve this, we will use the Bank Marketing dataset, which is available from the UCI Machine Learning Repository.

## Bank Marketing Dataset Overview:

The Bank Marketing dataset contains information related to marketing campaigns conducted by a bank to promote term deposits. Each record in the dataset represents a client and includes various features such as age, job, marital status, education, and more. The target variable, "y," indicates whether the client subscribed to a term deposit (binary: 'yes' or 'no').

### Problem Statement:

The primary objective of this project is to develop a predictive model that can assist the bank in identifying potential customers who are likely to subscribe to a term deposit. By accurately predicting customer behavior, the bank can optimize its marketing strategies and allocate resources more efficiently.

### Goals

Predict Customer Behavior: The primary goal is to develop an accurate predictive model that can determine whether a customer is likely to purchase a product or service offered by the bank. This prediction can assist the bank in targeting potential customers effectively.

Optimize Marketing Efforts: By accurately predicting customer behavior, the bank can optimize its marketing strategies. This includes allocating marketing resources more efficiently to focus on individuals who are more likely to subscribe to the bank's offerings.

Improve Campaign Efficiency: The model aims to improve the efficiency of marketing campaigns. It can help reduce marketing costs and increase the return on investment by targeting the right audience.

Enhance Decision-Making: The decision tree classifier provides interpretable results, allowing the bank to understand which features are most influential in predicting customer behavior. This information can guide future marketing decisions and product offerings.

## Data Cleaning

#### Checking for duplicates

```python
# A function to check for duplicates in our datasets
def identify_duplicates(df):
    """Simple function to identify any duplicates"""
    # identify the duplicates (dataframename.duplicated() , can add .sum() to get total count)
    # empty list to store Bool results from duplicated
    duplicates = []
    for i in df.duplicated():
        duplicates.append(i)
    # identify if there is any duplicates. (If there is any we expect a True value in the list duplicates)
    duplicates_set = set(duplicates) 
    if (len(duplicates_set) == 1):
        print("The Data has no duplicates")
    else:
        no_true = 0
        for val in duplicates:
            if (val == True):
                no_true += 1
        # percentage of the data represented by duplicates 
        duplicates_percentage = np.round(((no_true / len(df)) * 100), 3)
        print(f"The Data has {no_true} duplicated rows.\nThis constitutes {duplicates_percentage}% of the data set.") 



identify_duplicates(df)
```

#### Checking for missing values

```python
# A function to check for missing values in our dataset
def check_missing_values(data):
    # Count missing values in each column
    missing_values = data.isnull().sum()

    # Convert missing values count to percentage of total rows
    missing_percent = (missing_values / len(data)) * 100

    # Combine the missing values count and percent into a DataFrame
    missing_df = pd.concat([missing_values, missing_percent], axis=1)
    missing_df.columns = ['Missing Values', '% of Total']
# Return only columns with missing values
    missing_df = missing_df[missing_df['Missing Values'] > 0]

    return missing_df

# Check missing values in each dataset
display(check_missing_values(df))
```
